```python
import pandas as pd

# Membaca file CSV dengan delimiter ";"
df = pd.read_csv('/content/hasil_labelling.csv', delimiter=';')

# Menampilkan beberapa baris pertama
df.head()
```

```python
# Hapus kolom 'Unnamed'
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Tampilkan dataframe setelah kolom dihapus
df.head()
```

```python
#Pisahkan kolom full text dan sentiment
x = df['full_text']
y = df['Sentiment']
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Replace NaN values in 'full_text' column with empty strings
x = df['full_text'].fillna('')  # This line is changed

# Fit the TfidfVectorizer
tf_idf = TfidfVectorizer(ngram_range=(1,2))
tf_idf.fit(x)
```

```python
#Jumlah Fitur
print(len(tf_idf.get_feature_names_out()))
```

```python
print(tf_idf.get_feature_names_out())
```

```python
x_tf_idf = tf_idf.transform(x)
data_tf_idf = pd.DataFrame(x_tf_idf.toarray(), columns=tf_idf.get_feature_names_out())
data_tf_idf
```

```python
# Tambahkan ini di bagian awal kode Anda
import pickle

# Simpan objek x_tf_idf ke dalam file pickle
with open('tf_idf_feature.pickle', 'wb') as output:
    pickle.dump(x_tf_idf, output)
```

MODELLING SVM

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from datetime import datetime
start_time = datetime.now()
from joblib import dump
```

```python
#Proses memisahkan data uji dan data latih. Perbandingan 80% untuk data latih, 20% untuk data uji
X_train, X_test, y_train, y_test = train_test_split(X_kbest_features, y, test_size=0.2, random_state=42)
```

```python
print('Banyak data X_train:', len (X_train))
print('Banyak data X_test:', len (X_test))
print('Banyak data y_train:', len(y_train))
print('Banyak data y_test: ', len(y_test))
```

```python
# Mencari parameter terbaik dengan GridSearch
svm_pipeline = Pipeline([
    ('svm', SVC(class_weight='balanced'))
])

params = {
    'svm__kernel': ['linear', 'rbf'],
    'svm__C': [0.1, 10, 100],
    'svm__gamma': [0.5, 10, 100],
    'svm__decision_function_shape': ['ovr', 'ovo']
}

grid_search = GridSearchCV(svm_pipeline, param_grid=params, cv=3, n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
#Melatih model SVM dengan parameter terbaik
best_svm = grid_search.best_estimator_
best_svm.fit(X_train, y_train)
```

```python
#Cek parameter terbaik yang diperoleh dari GridSearch
print("Best Parameters:", grid_search.best_params_)
```

```python
#Simpan model SVM
dump(best_svm, filename="model_sentiment_svm.joblib")

end_time = datetime.now()
result_time =end_time-start_time
print("Duration:", result_time)
```

```python
#Prediksi label pada data uji
model_pred = best_svm.predict(X_train)
model_pred
```

MODELLING NAIVE BAYES

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from datetime import datetime
start_time = datetime.now()
from joblib import dump
```

```python
#Proses memisahkan data uji dan data latih. Perbandingan 80% untuk data latih, 20% untuk data uji
X_train, X_test, y_train, y_test = train_test_split(X_kbest_features, y, test_size=0.2, random_state=42)
```

```python
print('Banyak data X_train', len (X_train))
print('Banyak data X_test', len(X_test))
print('Banyak data y_train:',len(y_train))
print('Banyak data y_test', len(y_test))
```

```python
#Training model
algorithm = MultinomialNB (alpha=0.01, fit_prior =True)
model_nb = algorithm.fit(X_train, y_train)

#Simpan model hasil traning
dump (model_nb, filename='model_nb.joblib')
end_time = datetime.now()
result_time =end_time-start_time
print("Duration:", result_time)
```

```python
#Prediksi label pada data uji
model_pred_nb = model_nb.predict(X_train)
model_pred_nb
```

EVALUASI SVM

```python
#Jumlah data yang berhasil di prediksi & salah di prediksi
prediksi_benar = (model_pred_nb == y_train).sum() # Assign the sum to prediksi_benar
prediksi_salah = (model_pred_nb != y_train).sum() # Assign the sum to prediksi_salah

print('Jumlah prediksi benar\t:', prediksi_benar)
print('Jumlah prediksi salah\t:', prediksi_salah)

accuracy = prediksi_benar / (prediksi_benar + prediksi_salah)*100
print('Akurasi pengujian\t:', accuracy, '%')
```

```python
#Visualisasi prediksi benar dan salah SVM
import matplotlib.pyplot as plt
prediksi_benar = (model_pred_nb == y_train).sum()
prediksi_salah = (model_pred_nb != y_train).sum()
jumlah = [prediksi_benar, prediksi_salah]
labels = ['Prediksi Benar', 'Prediksi Salah']
fig, ax = plt.subplots()
rects = ax.bar(labels, jumlah, color=['#1A5F7A'])
for i, rect in enumerate (rects):
  height = rect.get_height()
  ax.text(rect.get_x() + rect.get_width()/2., height + 5, int(height),
          ha='center', va='bottom', color='black', fontweight='bold')
plt.xlabel('Prediksi')
plt.ylabel('Jumlah')
plt.show()
```

```python
#Confusion Matrix SVM
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_train, model_pred_nb)  # Change model_pred to model_pred_nb
print('Confusion matrix:\n', cm)
```

```python
#Classification Report SVM
from sklearn.metrics import classification_report
print('Classification report:\n',
      classification_report(y_train, model_pred_nb, digits=2, output_dict=False, target_names=['Negatif', 'Netral', 'Positif']))
```

```python
#Visualisasi confusion matrix SVM
import seaborn as sns
from sklearn.metrics import confusion_matrix
sns.heatmap(cm, annot=True, cmap='Blues', fmt='g')
plt.title('Confusion Matrix')
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.show()
```
